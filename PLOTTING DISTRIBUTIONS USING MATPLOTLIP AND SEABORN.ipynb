#### Basic statistical functions of the dataset.

```python
plt.boxplot(data['Total Sales'])
plt.show()
```

#### The boxplot shows that there are more outliers and lies in the range from 380000 to 630000. The sales had made at the minimum value of 1 to the maximum value of 823000. The mean lies at 93000 and maximum 75% of the sales happened at the value of 150000.

```python
plt.boxplot(data['Total Sales'])
plt.yscale('log')

plt.show()
```

#### Log has been taken for the above boxplot for better understanding above the data. 

```python
plt.hist(data['Price per Unit'])
plt.show()
```

#### This histogram describes about the price of the products. It shows that company selling more products for the price between 40 to 50. There are very less products for the price above 100. The company need to concentarte more on the products priced 60 and above.

```python
plt.scatter(data['Total Sales'], data['Operating Profit'])
plt.show()
```

#### The scatterplot shows the relationship between sales and the profit. It shows that majority of the sales had made below 280000 and also the sales had made for both profit and the negative profit. The outliers here describes that the sales for high price has taken place and also that sales has given high profit too.

```python
sns.distplot(data['Total Sales'])
plt.show()
```

#### The distplot shows that majority of the sales had taken between the values from 1 to 10000. There were very less sales for the value above 600000.

```python
sns.distplot(data['Total Sales'][:200],rug=True)
plt.show()
```

#### The rug plot describes above the sales. The density of the rug describes that the majority of the sales occurance. Here, the density of the rug are more between the value 115000 and 300000.

```python
sns.distplot(data['Total Sales'],hist=False)
plt.show()
```

#### This distplot is without histogram. Here the trend line tells about the sales happened. The trendline is peak at the value from 1 to 100000 which describes that sales had made high during that range.

```python
sns.boxplot(data['Price per Unit'])
plt.title('Price per Unit')

plt.show()
```

####  The boxplot tells about the price of the products. Here the mean lies at 45 and the minimum unit price for the product is 7 and the maximum value is 110. 50% of the prices falls between 35 to 55. 

```python
plt.subplot(3,2,1)
plt.title('Sales')
sns.distplot(data['Total Sales'])

plt.subplot(3,2,2)
plt.title('Unit Price')
sns.distplot(data['Price per Unit'])

plt.subplot(3,2,3)
plt.title('Units Sold')
sns.distplot(data['Units Sold'])

plt.subplot(3,2,4)
plt.title('Profit')
sns.distplot(data['Operating Profit'])

plt.tight_layout()
plt.show()
```

#### The subplot had created for the variables sales, unit price of the product, units sold and profit earned. Here, subplot is helped to compare the variables easily.

```python
sns.jointplot(x='Total Sales',y='Operating Profit',data=data,color='lightgreen')
plt.show()
```

#### The jointplot is created to describe the realtionship between total sales and the profit earned. The majority of the sales had taken place approximately between 9000 to 150000 and the profits are gained high below 50000. There are some outlier which shows that high priced products are sold less and that gave high profit also. 

```python
data1 = data[(data['Total Sales'] >185000) & (data['Operating Profit'] >50000)]

sns.jointplot(x='Total Sales',y='Operating Profit', data=data1, kind="hex", color='purple')
plt.show()
```

#### The heat map shows about the sales and profit relationship. The dark shade indicates that majority of the sales had taken place approximately between 9500 to 200000 and the profits are gained high between 50000 to 100000.
